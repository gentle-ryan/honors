```python
import pandas as pd
data = pd.read_csv("train.csv")
```

```python
import re
```

```python
data = data[['context', 'target', 'output']]
```

```python
data['output'] = data['output'].apply(lambda x: 0 if x == 'entailment' else 1)
```

```python
data
```

```python
def extract_korean_with_punctuation(text):
    return re.findall(r'[가-힣\s.?!\-\[\]()"\']+', text)

# 각 셀에서 한글 및 문장부호만 추출하여 새로운 컬럼으로 나누기
data['context'] = data['input'].apply(lambda x: ''.join(extract_korean_with_punctuation(eval(x)['premise'])))
data['target'] = data['input'].apply(lambda x: ''.join(extract_korean_with_punctuation(eval(x)['proposition'])))
```

```python
data2 = pd.read_csv("test.csv")
```

```python
data = pd.concat([data, data2], axis = 0)
```

```python
import json
import csv

def jsonl_to_csv(jsonl_file_path, csv_file_path):
    # jsonl 파일을 열고 각 줄을 읽어서 json 객체로 변환
    with open(jsonl_file_path, 'r', encoding='utf-8') as jsonl_file:
        json_list = [json.loads(line) for line in jsonl_file]
    
    # csv 파일에 쓰기
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        if not json_list:
            return  # json_list가 비어있다면 종료

        # csv 파일의 필드명 설정 (첫 번째 json 객체의 키를 사용)
        fieldnames = json_list[0].keys()
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        # 헤더 작성
        writer.writeheader()

        # json 객체들을 csv에 작성
        for json_obj in json_list:
            writer.writerow(json_obj)

# 사용 예시
jsonl_file_path = 'nikluge-te-2023-dev.jsonl'
csv_file_path = 'temp_test.csv'
jsonl_to_csv(jsonl_file_path, csv_file_path)
```
